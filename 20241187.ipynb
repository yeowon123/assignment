#1. Leetcode 225. Implement Stack using Queues

```python
from collections import deque


class MyStack(object):

    def __init__(self):
        self.q1 = deque()
        self.q2 = deque()

    def push(self, x: int) -> None:
        self.q2.append(x)
        while self.q1:
            self.q2.append(self.q1.popleft())
        self.q1, self.q2 = self.q2, self.q1

    def pop(self) -> int:
        return self.q1.popleft()

    def top(self) -> int:
        return self.q1[0]

    def empty(self) -> bool:
        return not self.q1
```

## 해설
- Stack은 LIFO 구조, Queue는 FIFO 구조
- push:
   - 새 원소를 'q2'에 삽입
   - 기존의 'q1' 원소들을 'q2'로 모두 옮긴 뒤 'q1','q2'를 swap
- 이렇게 하면 'q1'의 front가 Stack의 top 역할을 함
- 'pop','top','empty'는 모두 'q1'의 front 기준으로 처리

#2. Leetcode 232. Implement Queue using Stacks

```python
class MyQueue(object):

    def __init__(self):
        self.in_stack = []
        self.out_stack = []


    def push(self, x:int) -> None:
        self.in_stack.append(x)


    def pop(self) -> int:
        if not self.out_stack:
            while self.in_stack:
                self.out_stack.append(self.in_stack.pop())
        return self.out_stack.pop()


    def peek(self) -> int:
        if not self.out_stack:
            while self.in_stack:
                self.out_stack.append(self.in_stack.pop())
        return self.out_stack[-1]


    def empty(self) -> bool:
        return not self.in_stack and not self.out_stack
```

## 해설
- Queue는 FIFO, Stack은 LIFO
- 'in_stack' : 입력을 저장하는 스택
- 'out_stack' : 실제 pop/peek할 때 사용하는 스택
- 'pop','peek' 전에 'out_stack'이 비었으면 'in_stack'의 모든 요소를 꺼내 'out_stack'에 옮긴다
- 이렇게 하면 queue처럼 앞의 요소부터 꺼낼 수 있음


#3. 교재의 큐 연습문제 전부

```python
# 3-1
class ListQueue:
  def __init__(self):
    self.__queue = []

  def enqueue(self,item):
    self.__queue.insert(0,item)

  def dequeue(self):
    return self.__queue.pop()

  def front(self):
    if self.__queue.is_empty():
      return None
    else:
      return self.__queue[len(self.__queue)-1]

  def is_empty(self) -> bool:
    return not bool(self.__queue)

  def dequeue_all(self):
    self.__queue.clear()
```

```python
# 3-2
def is_in_set(string:str) -> bool:
  q = ListQueue()
  index = 0
  while string[index] != '$':
    q.enqueue(string[index])
    index += 1
  index += 1
  while index < len(string):
    if q.dequeue() != string[index]:
      return False
    else:
      index += 1
  return True
```

```python
# 3-3
def copy_linked_queue(a: LinkedQueue, b: LinkedQueue):
    b.dequeue_all()
    temp_queue = LinkedQueue()

    while not a.is_empty():
        temp_queue.enqueue(a.dequeue())

    while not temp_queue.is_empty():
        value = temp_queue.front()
        a.enqueue(value)
        b.enqueue(temp_queue.dequeue())
```

```python
# 3-4
def push(q1: LinkedQueue, q2: LinkedQueue, item):
    q2.dequeue_all()
    while not q1.is_empty():
        q2.enqueue(q1.dequeue())
    q1.enqueue(item)
    while not q2.is_empty():
        q1.enqueue(q2.dequeue())

def pop(q1: LinkedQueue, q2: LinkedQueue):
    q2.dequeue_all()
    count = 0
    while not q1.is_empty():
        q2.enqueue(q1.dequeue())
        count += 1
    for _ in range(count - 1):
        q1.enqueue(q2.dequeue())
    return q2.dequeue()
```

```python
# 3-5
def enqueue(s1: LinkedStack, s2: LinkedStack, item):
    s2.pop_all()
    while not s1.is_empty():
        s2.push(s1.pop())
    s1.push(item)
    while not s2.is_empty():
        s1.push(s2.pop())

def dequeue(s1: LinkedStack, s2: LinkedStack):
    s2.pop_all()
    while not s1.is_empty():
        s2.push(s1.pop())
    ret_item = s2.pop()
    while not s2.is_empty():
        s1.push(s2.pop())
    return ret_item
```
