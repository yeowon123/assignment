```python
from collections import deque


class MyStack(object):

    def __init__(self):
        self.q1 = deque()
        self.q2 = deque()

    def push(self, x: int) -> None:
        self.q2.append(x)
        while self.q1:
            self.q2.append(self.q1.popleft())
        self.q1, self.q2 = self.q2, self.q1

    def pop(self) -> int:
        return self.q1.popleft()

    def top(self) -> int:
        return self.q1[0]

    def empty(self) -> bool:
        return not self.q1
```

```python
## 해설
- Stack은 LIFO 구조, Queue는 FIFO 구조
- push:
   - 새 원소를 'q2'에 삽입
   - 기존의 'q1' 원소들을 'q2'로 모두 옮긴 뒤 'q1','q2'를 swap
- 이렇게 하면 'q1'의 front가 Stack의 top 역할을 함
- 'pop','top','empty'는 모두 'q1'의 front 기준으로 처리
```
